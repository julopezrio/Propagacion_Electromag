```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Parámetros físicos (puedes cambiarlos)
epsilon_0 = 8.85e-12  # Permitividad del vacio (F/m)
mu_0 = 1.26e-6        # Permeabilidad del vacio (H/m)

amplitud = 1
frecuencia = 150e6    # Frecuencia en Hz
phi=0   #desfase


medio=["vacio","papel","FR4","cobre"]
epsilon_r=[1,3.5, 4.5, 8.85e-12 ]
mu_r=[1,1,1,0.999994 ]

dic=np.array([medio,epsilon_r,mu_r]) #diccionario de valores de permitividad y permeabilidad

# Dominio espacial y temporal
t = np.linspace(0, 1e-9, 1000)  # metros
x = np.linspace(0, 100000000, 1000)
X, T = np.meshgrid(x, t)

for i in range(len(medio)):
    dic[:,i]
    velocidad = 1 / np.sqrt(mu_0*float(dic[2:,i])* epsilon_0*float(dic[1:2,i]))
    y=amplitud* np.sin(2 * np.pi * frecuencia * ((X/velocidad)-(T)) + phi)

    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, T, y, cmap='viridis')
    ax.set_title(dic[0:1,i])
     # Etiqueta con el valor de phi

plt.xlabel('t')
plt.ylabel('y')
plt.grid()
#plt.title('sen(x+phi)')
plt.legend()


ax.set_xlabel("Posición (x) [m]")
ax.set_ylabel("Tiempo (t) [s]")
ax.set_zlabel("Amplitud")

plt.show()

```

```python

xx = np.linspace(0,4, 1000)

for i in range(len(medio)):
    velocidad = 1 / np.sqrt(mu_0*float(dic[2:,i])* epsilon_0*float(dic[1:2,i]))
    y1=amplitud*np.sin(2*np.pi*xx*frecuencia*(1/velocidad))
    plt.plot(xx, y1, label=f'{dic[0:1,i]}')  # Etiqueta con el valor de phi

plt.xlabel('x')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```
